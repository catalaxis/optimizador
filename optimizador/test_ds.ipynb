```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from optimizador.data_loader import DataLoader
from optimizador.optimizer import Optimizer
from optimizador.result import ResultsHandler
```

```python
## Cargar optimization_problem_data.csv
def load_optimization_data(file_path):
    data_loader = DataLoader(file_path)
    data_loader.load_data()
    return data_loader.get_data()

# Example usage:
data = load_optimization_data('optimization_problem_data.csv')
optimizer = Optimizer(**data)
result = optimizer.solve()
print(f"Optimal production: Product A = {result[0]}, Product B = {result[1]}")
print(f"Maximum profit: {result[2]}")
```

```python
## Grafico de resultados
results_handler = ResultsHandler(result[0], result[1], result[2])
results_handler.plot_results()
```
